## 2.  Questions & Answers

### Q1: What is Artificial Intelligence (AI)?
**Answer:** Artificial Intelligence (AI) is a branch of computer science focused on creating machines or software capable of thinking, understanding, learning, and making decisions just like humans. Its primary goal is to make computers intelligent so they can solve complex problems autonomously.

### Q2: What is Machine Learning (ML)?
**Answer:** Machine Learning (ML) is a sub-field of AI. Instead of using explicit programming (writing code for every single line) we train computers to learn directly from data. ML algorithms analyze historical data, identify underlying patterns, and make future predictions or decisions based on that analysis. As the system processes more data, its accuracy improves continuously.

### Q3: What is Deep Learning (DL)?
**Answer:** Deep Learning (DL) is a specialized and advanced subset of Machine Learning inspired by the structure and function of the human brain (Neural Networks). It utilizes "Deep Neural Networks" that can automatically extract and learn features from massive, complex datasets, such as images, audio, and unstructured text.

### Q4: What is the difference between AI, ML, and DL?
**Answer:** The relationship between these three terms can be understood as follows:
*   **AI (Artificial Intelligence):** The broad umbrella concept encompassing any technology that enables a machine to mimic human intelligence.
*   **ML (Machine Learning):** A subset of AI focused on using algorithms to allow systems to learn from data on their own.
*   **DL (Deep Learning):** A further subset of ML that uses multi-layered neural networks to handle highly complex data patterns.
*(In simple terms: DL is a subset of ML, and ML is a subset of AI).*
