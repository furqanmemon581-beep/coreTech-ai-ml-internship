```python
# Example 1: Primitive Data Types (String, Integer, Float, Boolean)
name = "Alice"          # String
age = 28               # Integer
height = 5.6           # Float
is_student = False     # Boolean

print("--- Example 1: Primitives ---")
print(f"Name: {name} ({type(name)})")
print(f"Age: {age} ({type(age)})")
print(f"Height: {height} ({type(height)})")
print(f"Is Student: {is_student} ({type(is_student)})")
```

```python
# Example 2: Collection Data Types (List, Tuple, Dictionary)
skills = ["Python", "SQL", "Tableau"]      # List (Mutable, ordered)
coordinates = (40.7128, -74.0060)          # Tuple (Immutable, ordered)
user_profile = {"id": 101, "role": "Admin"} # Dictionary (Key-Value pairs)

print("\n--- Example 2: Collections ---")
print(f"Skills List: {skills} ({type(skills)})")
print(f"Coordinates Tuple: {coordinates} ({type(coordinates)})")
print(f"User Profile Dict: {user_profile} ({type(user_profile)})")
```

```python
# Example 3: If-Elif-Else Conditionals
score = 85

print("--- Example 3: Conditionals ---")
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C")
```

```python
# Example 4: For Loop (Iterating over a list)
print("\n--- Example 4: For Loop ---")
languages = ["Python", "R", "Julia"]
for lang in languages:
    print(f"I love coding in {lang}")
```

```python
# Example 5: While Loop
print("\n--- Example 5: While Loop ---")
count = 3
while count > 0:
    print(f"Countdown: {count}")
    count -= 1
```

```python
# Example 6: Basic Function with Return Value
def calculate_bmi(weight_kg, height_m):
    return round(weight_kg / (height_m ** 2), 2)

print("--- Example 6: Basic Function ---")
bmi = calculate_bmi(70, 1.75)
print(f"Calculated BMI: {bmi}")
```

```python
# Example 7: Function with Default Arguments
def greet_user(username="Guest"):
    return f"Welcome back, {username}!"

print("\n--- Example 7: Default Arguments ---")
print(greet_user("Sarah"))
print(greet_user())
```

```python
# Example 8: Simple List Comprehension (Squaring numbers)
numbers = [1, 2, 3, 4, 5]
squares = [num ** 2 for num in numbers]

print("--- Example 8: List Comprehension ---")
print(f"Original: {numbers}")
print(f"Squares:  {squares}")
```

```python
# Example 9: Conditional List Comprehension (Filtering even numbers)
even_squares = [num ** 2 for num in numbers if num % 2 == 0]

print("\n--- Example 9: Conditional List Comprehension ---")
print(f"Squares of even numbers: {even_squares}")
```

```python
import numpy as np

# Example 10: Creating NumPy Arrays and Array Operations
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

print("--- Example 10: NumPy Array Operations ---")
print(f"Addition:       {arr1 + arr2}")
print(f"Multiplication: {arr1 * 2}")
```

```python
# Example 11: NumPy Statistical Functions
data_points = np.array([12, 15, 18, 22, 30, 45])

print("\n--- Example 11: NumPy Statistics ---")
print(f"Mean:   {np.mean(data_points)}")
print(f"Max:    {np.max(data_points)}")
print(f"Std Dev:{round(np.std(data_points), 2)}")
```

```python
import pandas as pd

# Example 12: Creating a Sample Dataset using a Dictionary
raw_data = {
    "Employee_ID": [1001, 1002, 1003, 1004, 1005],
    "Name": ["John", "Anna", "Peter", "Linda", "Tom"],
    "Department": ["HR", "Engineering", "Marketing", "Engineering", "Finance"],
    "Salary": [55000, 85000, 62000, 91000, 72000],
    "Experience_Yrs": [2, 5, 3, 7, 4]
}

print("--- Example 12: Dictionary Dataset ---")
print(raw_data)
```

```python
# Example 13: Converting Dictionary to Pandas DataFrame & Displaying DataFrame
df = pd.DataFrame(raw_data)

print("\n--- Example 13: Pandas DataFrame ---")
print(df)
```

```python
# Example 14: Displaying Basic Info and Metadata
print("\n--- Example 14: DataFrame Info ---")
df.info()
```

```python
# Example 15: Descriptive Statistics
print("\n--- Example 15: Descriptive Statistics ---")
print(df.describe())
```
