```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
data = {
    "ClientID": [1, 2, 3, 4, 4],
    "Name": ["Ali", "Sara", "Ahmed", "Zain", "Zain"],
    "Age": [25, 30, None, 28, 28],
    "City": ["Karachi", "Lahore", "Karachi", "Islamabad", "Islamabad"],
    "Income": [50000, 60000, 55000, None, 65000],
    "Gender": ["Male", "Female", "Male", "Male", "Male"],
    "Purchase": ["Yes", "No", "Yes", "No", "No"]
}

df = pd.DataFrame(data)

df
```

```python
df = df.drop_duplicates()

df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Income"].fillna(df["Income"].mean(), inplace=True)

df
```

```python
le = LabelEncoder()

df["Gender"] = le.fit_transform(df["Gender"])
df["Purchase"] = le.fit_transform(df["Purchase"])

df = pd.get_dummies(df, columns=["City"])

df
```

```python
scaler = MinMaxScaler()

df[["Age", "Income"]] = scaler.fit_transform(df[["Age", "Income"]])

df
```

```python
df_final = df.drop(["Name"], axis=1)

df_final
```

```python
df_final.to_csv("cleaned_data.csv", index=False)
print("File saved successfully!")
```
